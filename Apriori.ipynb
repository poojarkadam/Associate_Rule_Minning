```python
import numpy as np
import pandas as pd
from apyori import apriori
```

```python
store_data = pd.read_csv('store_data.csv', header=None)
store_data.head()
```

```python
store_data.shape
```

```python
store_data_list =  [[str(store_data.values[i,j]) for j in range(store_data.shape[1])]  for i in range(store_data.shape[0])]
print(len(store_data_list))
print(store_data_list[0])
```

```python
results = list(apriori(store_data_list, min_support=0.005))
print(len(results))
```
